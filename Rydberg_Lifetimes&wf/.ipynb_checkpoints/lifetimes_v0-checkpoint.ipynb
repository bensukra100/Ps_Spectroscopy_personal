```python
## importing modules 
import numpy as np
import matplotlib.pyplot as plt
from scipy.constants import elementary_charge, m_e, c, hbar, alpha, Rydberg, epsilon_0, h
from tqdm import tqdm
from positronium.constants import lifetime_oPs
from positronium import Ps
import sympy as sympy
import scipy as scipy
from scipy.special import factorial
from wigner_3j import *
import matplotlib as mpl
from mpl_toolkits.axes_grid1 import make_axes_locatable
from matplotlib.ticker import MaxNLocator
from matplotlib.ticker import MultipleLocator
markers = mpl.markers.MarkerStyle.filled_markers

## matplotlib text/marker features
mpl.rc('font',family='serif')
fsize = 23
lnwidth = 1
tickwidth = 1
msize = 8

## defining and renaming some variables
Ryd_Ps = Rydberg/2
e = elementary_charge
a_0 = hbar / (m_e * c * alpha)
a_Ps = 2.0 * a_0
E_hPs = 2 * h * c *Ryd_Ps

%matplotlib inline
```

The radiative lifetime of a Rydberg state in some defined $\ell$-state is given by the inverse of the decay rates to all allowed decay pathways to lower-lying states $\ket{n' \ell'}$, 
$$\tau_{n\ell} = \Gamma_{n\ell}^{-1} = \Big[ \Sigma_{n'\ell'}A_{n'\ell', n \ell}\Big]^{-1}$$

The Einstein-$A$ coefficients are given by,

$$A_{n'\ell',n\ell} = \frac{2 e^2 \omega^3_{n'\ell',n\ell}}{3 \epsilon_0 h c^3} \frac{\ell_{max}}{2\ell +1}|\bra{n'\ell'}r\ket{n\ell}|^2,$$

where $\omega_{n'\ell',n\ell}$ is the angular frequency corresponding to the energy difference between the states, and the other constants are as usually defined.

### Note: 
I write n_2 as the higher/excited/initial state and n_1 as the lowwr/de-excited state. 

el variables are for l since l looks like 1.

```python
## the limits on the radial integrals, you can go further but then consider the resolution
r_0, r_T = 0, 2000*a_Ps 
N = 10000
rs = np.linspace(r_0, r_T, N) 
dr = (r_T-r_0)/N


def Ps_R_nl(r,n, el):
    '''function for determining the amplitude of the Ps radial wavefunction at a point r for some n and l'''
    Z = 1.
    N_nl = ((2.*Z/(n*a_Ps))**(3./2.))*np.sqrt(scipy.math.factorial(n-el-1.)/((2.*n)*scipy.math.factorial(n+el)))
    A = (((2.*Z*r)/(n*a_Ps))**el)*np.exp(-(Z*r)/(n*a_Ps))
    B = scipy.special.genlaguerre(n-el-1.,2.*el+1.)((2.*Z*r)/(n*a_Ps))
    return N_nl*A*B

def radial_integral(r_0, r_T, n_2, el_2, n_1, el_1):
    '''calculates the radial integral/ overlap of two radial wave functions from r_0 to r_T'''
    a = np.multiply(np.conj(Ps_R_nl(rs, n_2, el_2)),Ps_R_nl(rs, n_1, el_1))
    integral = np.sum(np.multiply(a, np.power(rs,3)))*dr
    return integral

def E_n(n):
    '''calculates energy of state n'''
    E = -1.*E_hPs/(2.*n**2.)
    return E

def transition_freq(n_2, n_1):
    '''calculates transition frquency of n2 to n1 in 0 E-field'''
    dE = E_n(n_2) -E_n(n_1)
    freq = (dE)/h
    return freq

def Einstein_A(n_2, el_2, n_1, el_1):
    '''calculates the Einstein A coefficient for a transition from n_2, el_2 to n_1, el_1
    Note: this expression has already been simplified to account for transition rules: 
    (1) del(el) = +/- 1
    (2) del(m)  = 0
    '''
    freq = transition_freq(n_2, n_1)*2.*np.pi
    overlap = radial_integral(0, r_T, n_2, el_2, n_1, el_1)
    A = ((2.*e**2.*(freq)**3.)/(3.*epsilon_0*h*c**3.))*(max(el_2,el_1)/(2.*el_2+1.))*overlap**2.
    return A

def fluor_lf(A_array):
    '''calculates radiative lifetime to ground state from excited n,l '''
    lf = (np.sum(A_array))**(-1.)
    return lf 
```

```python
el=1
el = [0,1,2,00] ##s,p,d, placeholder <- I don't know what circular state is until I know n_2
lifetime = [] ##array for holding lifetimes
n2s = np.arange(10, 20+1, 1) ##range of initial n_2's

for el_2 in range(len(el)): ##through the considered el_2's
    lifetime.append([]) ##
    for n_2 in tqdm(n2s): ## n's being considered
        el = [0,1,2,n_2-1] ##for the initial el_2's, notice circular state is being calculated
        A = 0
        for n_1 in range(1, n_2, 1): ##all posible n_1 the excited state can fluoresce to
            for el_1 in range(0, n_1, 1): ## over all l_1 in that n_1
                if el_1 == el[el_2] +1 or el_1 == el[el_2] -1: ##selection rule
                    A = A + Einstein_A(n_2, el[el_2], n_1, el_1) 
                else: A = A 
        lifetime[el_2].append(A**-1.) 
```

In a stray or weak background electric field, Rydbergs states have a mixed-$\ell$ character and states have radiative lifetimes that lie between the short-lived, low-$\ell$ states and the lond-lived, high-$\ell$ states. In these cases the Rydberg states are instead labelled by the parabolic quantum numbers $n_1$ and $n_2$ (or $k$ where $|k| = |n_1 - n_2|$). The lifetimes of these $\ell$-mixed states ($\bra{n n_1 n_2 m}$) can be determined by a weighter sum of the lifetimes of the $\bra{n\ell}$ components which comprise the the $\bra{n n_1 n_2 m}$ states. The weight is the square of the transformation coefficients, $\braket{n n_1 n_2 m}{n\ell m}$ which can be evaluated using the Wigner-3j coefficients, 
$$\braket{n n_1 n_2 m}{n\ell m} = (-1)^{(\ell - n + m +n_1-n_2)-2 + \ell} \sqrt{2\ell+1}\begin{pmatrix}
\frac{n-1}{2} & \frac{n-1}{2} & \ell \\
\frac{m+n_1-n_2}{2} & \frac{m+n_1-n_2}{2} & -m
\end{pmatrix},$$

where the last term is a Wigner-3J symbol. The radiative lifetime of a $\ket{n n_1 n_2 m}$ state is therefore given by, 
$$\tau_{n n_1 n_2 m} = \Gamma_{n n_1 n_2 m} = \Big[ \Sigma_{n\ell}{|\braket{n n_1 n_2 m}{n\ell m}|^2\Gamma_{n\ell}}\Big]^{-1}$$

```python
## defining a function to calculate the Angular integrals using the Wigner 3-J shortcut 
def angular_integral(el_2, m_2, el_1, m_1):
    A = ((-1.)**(m_1))*np.sqrt((4.*np.pi)/(3.))*np.sqrt((3.*(2.*el_1+1.)*(2.*el_2+1.))/(4.*np.pi))
    W1 = wigner_3j(el_1, 1., el_2, -m_1, 0., m_2)
    W2 = wigner_3j(el_1, 1., el_2, 0., 0., 0.)
#     return A,W1,W2,A*W1*W2
    return A*W1*W2

def transition_dipole(r_0, r_T, n_2, el_2,m_2, n_1, el_1, m_1):
    N = 10000
    rs = np.linspace(r_0, r_T, N) 
    dr = (r_T-r_0)/N
    a = np.multiply(np.conj(Ps_R_nl(rs, n_2, el_2)),Ps_R_nl(rs, n_1, el_1))
    rad_integral = np.sum(np.multiply(a, np.power(rs,3)))*dr
    ang_integral = angular_integral(el_2, m_2, el_1, m_1)
    return rad_integral*ang_integral*e

def transformation_coefs(n, k, m, el):
    '''calculates the transformation coefficients for some nl state'''
    power = (1. - n + m + k)/2.+el
    a = (-1.)**(power)
    b = np.sqrt(2.*el + 1.)
    j_1 = (n-1.)/2.
    j_2 = (n-1.)/2.
    j_3 = el
    m_1 = (m+k)/2.
    m_2 = (m-k)/2.
    m_3 = -m
    c = wigner_3j(j_1, j_2, j_3, m_1, m_2, m_3)

    res = a*b*c
    return res
```

```python
lifetime = []
n_2s = [10,20]
i,j = 0,0

for n_2 in n_2s:
    lifetime.append([])
    j = 0
    for m in [0,1,2]:
        lifetime[i].append([])
        
        for k in tqdm(np.arange(-(n_2-m-1), (n_2-m-1)+1,2)):
            A = 0
            for el_2 in range(0,n_2, 1):
                for n_1 in np.arange(1, n_2):
                    for el_1 in np.arange(0, n_1, 1):
                        if (el_1 == el_2 + 1) or (el_1 == el_2 - 1):
                            A = A + Einstein_A(n_2, el_2, n_1, el_1)*(transformation_coefs(n_2,k,m,el_2)**2.)
#                             A = A + Einstein_A(n_2, el_2, n_1, el_1)
                        else: 
                            A = A 
            lifetime[i][j].append((A)**-1.)
        j +=1        
    i += 1
```
